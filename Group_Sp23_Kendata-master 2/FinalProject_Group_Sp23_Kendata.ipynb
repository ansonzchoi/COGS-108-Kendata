## Background & Prior Work

Kendo is a traditional Japanese martial art that uses bamboo swords and armor to spar with others. The Kendo Club at UCSD was founded in 1998 and promotes awareness and appreciation of this art by offering weekly practices through the UCSD Recreation Department [1]. The topic of interest in joining The Kendo Club at UCSD for UCSD undergraduate students is of interest to us because, as Kendo club members, we want to research which student demographics are most likely to be interested in Kendo. Through this project, we can better understand how to promote the club and increase member retention by appealing to the projected demographic(s). There has been no prior work done, to our knowledge, on the demographics of the average university student that participates in Kendo—Kendo is a relatively small community compared to other sports that student bodies commit to. However, there is data available on the undergraduate student population of UCSD that can help us to understand the overall demographics of UCSD. For example, the majority of undergraduate students at UCSD are Social Science majors (36.6%), followed by Engineering majors (18.8%), and Biology majors (17.5%) as of the 2021-2022 school year. Furthermore, the percentage of male students is slightly lower at 48.1% than female students at 50.7% [2]. We will take this information into consideration as potential biases when analyzing our data. We also used data on the distribution of racial identities and gender identities from the Department of Education for comparison with the data we collected ourselves. This data was collected nationally from 15.4 million undergraduate students enrolled in the fall of each year. In 2021, the Department of Education reported the highest number of undergraduate enrollment by race/ethnicity and nonresident status was students who identified as White (7.8 million) followed by students who identified as Hispanic (3.3 million) [3]. Furthermore, the data shows that 8.9 million female-identifying students and 6.5 male-identifying students enrolled in 2021 [3]. With this information, we can compare the large data set provided by the Department of Education to our collected data from Kenshi around the nation and perform a more accurate analysis of the effect different variables have on interest in Kendo. This background data we reviewed can be used as context for the data we collect as well as to perform data analysis tests.

# Data Cleaning

1) We first went through our survey responses and identified responses that were filled out incorrectly. Through code, we deleted these rows of our responses (in which there were 2 of them.)

2) We dropped the timestamps that were collected for each response as well since that information was automatically collected by Google Forms and was unnecessary.

3) We adjusted the title of our survey to ‘Kendo Demographic Survey’ and the survey’s questions were simplified to be easier to read and more accurately depict the information in the columns. For example, we changed the question ‘If you attended college, what major did you study?’ to ‘Major’.

4) We also observed that our question, “What college did you attend?”, had lots of variation. Our dataset included responses from 11 different colleges but ended up with 29 variations of the colleges’ names due to spelling and capitalization differences and the use of acronyms. To clean this data up, we mainly used the rename, drop, and replace functions to standardize this information down to the original 11 colleges.

5) We then ran into a similar issue with the majors that our respondents are pursuing/have completed. We found that it may be too difficult to code for every variation of specific majors to categorize them into their broader majors, so we manually transformed specific majors into their broader, less specific subjects. For example, we simplified ‘Structural Engineering’ and ‘Computer Engineering’ into just ‘Engineering’. And we simplified ‘Cognitive Science’ and ‘Chemistry’ into just ‘Science’.

6) We implemented the last strategy of manually cleaning our responses by applying them to the ‘occupation’ column of our non-student participants. The answers were far too specific and varied, so we simplified occupation titles down to the industry they were in. For example, we categorized ‘Optometrist’ and ‘Orthopedic Surgeon’ as ‘Doctor’ and categorized ‘Sales’ and ‘Management’ as ‘Business’. We also incorporated an ‘Not Working’ category for those who were not working or retired.

7) We also cleaned the “Grade Started(current students)” and “Grade Started(non-students)” columns by changing the non-null valuables from string inputs like “First Year” to “1” and “Third Year” to “3” so that we would be able to find the distribution in college years that indicate when people tend to join Kendo.

8) We then replaced “Yes”/“No” with Boolean “True”/“False,” respectively, when asking people if they are in college. By doing this, having Boolean values makes it easier to read and pass logical expressions.

# Ethics & Privacy

In conducting our data science project, we recognize the importance of addressing potential social and unintended consequences that may arise from our research. Our primary objective is to ensure that our data collection and analysis are conducted ethically and responsibly, taking into account the diverse perspectives and impacts that may be associated with our study.

Participants will be invited to complete the survey voluntarily and based on their own will and consent, with the freedom to discontinue the survey at any point without facing any negative consequences. However, we acknowledge the possibility of collection bias due to the distribution of the survey through Discord, as members who are more active on the platform may have a higher likelihood of participation compared to those who are less active.

While we will make efforts to mitigate this bias, we recognize that it may introduce unintended consequences, such as potential variations in interest among different demographics in the Kendo Club. If our findings were to reveal that Kendo was dominated by East Asian individuals, what impact would that have on non-East Asian interest and membership? Would our data discourage certain demographics from joining? To prevent this concern, we will keep our raw data confidential and only in the scope of this project. We will also not use this data to maliciously exclude certain communities from joining Kendo and rather use this data, as members of the Kendo board, to see what practices we can implement in the future to promote a more diverse club.

It is important to note that in addition to members, our survey data includes participants from other schools and alumni. We acknowledge that we cannot regulate how the survey was sent out to these participants, which may introduce additional biases and variations in responses. We will take this into consideration during our analysis and interpretation of the data.

To protect the privacy and confidentiality of our participants, we will limit the exposure of personally identifiable information (PII). Participants will be asked to provide only their names and will not be required to provide any additional identifying information such as PID. Furthermore, we will ensure that the survey questions are directly relevant to our research variables, avoiding any unnecessary intrusion into personal matters. The purpose and intended use of the collected data will be transparently communicated to all survey participants, allowing them to make informed decisions regarding their participation.

With regard to data security, we emphasize that the data collected will be used solely for academic purposes. The raw data obtained from the survey, hosted on Google Forms, will be accessible only to authorized team members. We will maintain the confidentiality of the data to the best of our ability and handle it in accordance with ethical data management practices.

Participants will have the right to withdraw their data at any time during the study, and we will explicitly inform them of this option within the survey itself. We are committed to respecting the autonomy of our participants and ensuring that their choices regarding data withdrawal are honored.

Additionally, we acknowledge that the ongoing COVID-19 pandemic may impact the data regarding the year in which students started Kendo. Given the restrictions on physical campus presence and limited club promotion activities during this time, it is possible that the interest and engagement of current second and third-year students in Kendo clubs may have been influenced. When analyzing our data, we will carefully consider the potential effects of the pandemic on students' ability to join the club, particularly during the specified time frame.

In summary, we are committed to conducting our data science project with a strong ethical foundation, ensuring that we address social and unintended consequences associated with our research. We uphold the principles of voluntary participation, privacy protection, data security, and transparent communication to uphold the integrity of our study and promote responsible data science practices.


# Conclusion & Discussion

In our analysis, we gained valuable insights into the correlation between demographic variables and the likelihood of individuals choosing to pursue kendo. We identified several trends based on the data we collected and other relevant data sets.

Firstly, we discovered that race significantly impacts Kendo participation, with certain racial groups being over- or under-represented in the practice. While Asians make up the smallest percentage of the population in the United States compared to other racial and ethnic groups, our data shows that individuals that identify as East Asian/Pacific Islander are the overwhelming majority in the kendo community. As Kendo has deep roots in East Asian culture and has been traditionally practiced and valued in countries like Japan, Korea, and China, these factors could reflect the higher likelihood of East Asians joining Kendo. 

For our second variable, gender identities, we found that most people who participate in Kendo are male identifying, even though the national enrollment data shows that most college students identify as female. This contrast indicates that kendo could be more appealing to male identifying individuals, possibly due to the fact that Kendo has historically been a male dominated martial art, and women were traditionally encouraged to participate in Naginata, an alternative martial art. Kendo has only become more inclusive in the recent decades. The result disproves our hypothesis that there would be no correlation between gender identities and participation in Kendo.


For our third variable, year started in college, we found that the majority of people started Kendo in their first year in college, followed by starting in the third year. This partially aligns with our hypothesis that there is a negative correlation between years started in college and participation in Kendo as most people did start in their first year, but the second most amount of people started in the third year. This could be attributed to students who transferred into college, as they are more inclined to explore new interests in their first year of college after transferring as third year students.


Lastly, we observed different majors and occupations that were more popular and less popular with those that participate in Kendo. Hypothetically, these variables’ categories would be evenly distributed but our data actually showed a statistically significant favoritism towards certain majors and occupations. Those enrolled in science majors were the most popular and the least popular majors were art, soft sciences, and undeclared. With the exception of business economics majors (which was the third most popular major), we found that the top four majors were all STEM oriented (science, engineering, computer science, math, respectively.) This aligns with the data we collected concerning careers that non-college students are in. We found that people who had careers as engineers and in business were the most popular. And the next popular occupations, such as doctors and scientists, were less than half as popular as the prior career paths. We believe this could be attributed to the fact that Kendo is an expensive sport to participate in—since the equipment and manufacturers are more niche and specialized—and the majors and occupations that were most observed are in notably lucrative fields.


Overall, this project has granted us an empirical sense of the variety of factors that determine what kind of members are more likely to participate in Kendo. We plan on leveraging our insights by sharing the results of our findings to other clubs and dojos across the United States that assisted in participating by filling out our survey. We hope that through our findings, we’ll be able to garner a wider diversity in Kendo and address accessibility issues—such as race, sex, age, and financial burdens—that bar individuals from exploring this exciting martial art.
